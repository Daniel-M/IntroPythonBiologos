# Graficación con Matplotlib y Numpy

## Descripción
Para este ejemplo se importan los módulos `matplotlib.pyplot` y `numpy` para generar las parejas de puntos $x,y$ que serán graficados al invocar el método plot.

```python
%matplotlib inline # Esta línea es necesaria para que las gráficas se muestren en el notebook

import numpy as np # importa numpy bajo el nombre de np
import matplotlib.pyplot as plt # importa matplotlib.pyplot bajo el nombre plt
```

```python
# Se genera conjunto numérico de valores x entre 0 y 3*pi con 500 puntos
x = np.linspace(0, 3*np.pi, 500)

# se crea la gráfica de los valores de x vs sin(x^2) mediante la función sin de numpy
plt.plot(x, np.sin(x**2))

# Se añade título a la gráfica
plt.title('Gráfica de $\sin(x^2)$')

#finalmente se muestra la gráfica
plt.show()
```

```python
# Se genera conjunto numérico de valores x entre -2*pi y 2*pi con 500 puntos
x = np.linspace(-2*np.pi, 2*np.pi, 500)

# se crea la gráfica de los valores de x vs sin(x^2)/x^2 mediante la función sin de numpy
plt.plot(x, np.sin(x**2)/x**2)

# Se añade título a la gráfica
plt.title('Gráfica de $\sin(x^2)/x^2$')

#finalmente se muestra la gráfica
plt.show()
```
